```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi çekme işlemi
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
response = requests.get(URL)

if response.status_code == 200:
    dataset_part_2_csv = response.content
    df = pd.read_csv(io.BytesIO(dataset_part_2_csv))
else:
    print("Veri çekme işlemi başarısız oldu. HTTP status kodu:", response.status_code)

# 'Orbit' sütunu üzerinde gruplama yapma ve başarı oranlarını hesaplama
orbit_success_rate = df.groupby('Orbit')['Class'].mean().reset_index()

# Sütun grafiği (bar chart) ile görselleştirme
plt.figure(figsize=(10, 6))
plt.bar(orbit_success_rate['Orbit'], orbit_success_rate['Class'])
plt.xlabel('Yörünge Türü')
plt.ylabel('Başarı Oranı (Ortalama)')
plt.title('Yörünge Türlerine Göre Başarı Oranları')
plt.ylim(0, 1)  # Başarı oranları 0 ile 1 arasındadır
plt.xticks(rotation=45)  # Yörünge türü etiketlerini 45 derece döndür
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Saçılma noktası grafiği (scatter plot) ile görselleştirme
sns.scatterplot(x='FlightNumber', y='Orbit', hue='Class', data=df)
plt.xlabel('Flight Number')
plt.ylabel('Orbit')
plt.title('Flight Number vs Orbit (Scatter Plot)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Saçılma noktası grafiği (scatter plot) ile görselleştirme
sns.scatterplot(x='PayloadMass', y='Orbit', hue='Class', data=df)
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Orbit')
plt.title('Payload Mass vs Orbit (Scatter Plot)')
plt.show()
```

```python
# Tarihten yıl bilgisini çıkaran fonksiyon
def Extract_year(date):
    return date.split("-")[0]

# "Date" sütunundaki her tarihi yıla dönüştürün
df['Year'] = df['Date'].apply(Extract_year)

# Yıllık ortalama başarı oranlarını hesaplayın
yearly_success_rate = df.groupby('Year')['Class'].mean().reset_index()

# Çizgi grafiği (line chart) ile görselleştirme
plt.figure(figsize=(10, 6))
plt.plot(yearly_success_rate['Year'], yearly_success_rate['Class'], marker='o', linestyle='-')
plt.xlabel('Yıl')
plt.ylabel('Ortalama Başarı Oranı')
plt.title('Yıllık Ortalama Başarı Trendi')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# İlgilendiğimiz değişkenleri seçme
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]

# Özellikler veri çerçevesini görüntüleme (ilk 5 satır)
features.head()
```

```python
# Kategorik sütunlara One-Hot Encoding uygulama
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Sonuçları görüntüleme (ilk 5 satır)
features_one_hot.head()
```

```python
# Tüm sayısal sütunları "float64" veri türüne dönüştürme
features_one_hot = features_one_hot.astype('float64')
```
